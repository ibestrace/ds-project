**2、动量因子测试**

```python
import pandas as pd
import numpy as np
import akshare as ak

# 选取上证A股指数成分股
stock_info = ak.stock_info_a_code_name()
tickers = stock_info['code'].tolist()
tickers
```

```python
# 计算动量因子
def momentum_factor(tickers, period='12mo'):
    close_prices = pd.DataFrame()
    for ticker in tickers:
        try:
            prices = ak.stock_zh_a_daily(symbol=ticker, adjust="hfq", start_date=(pd.Timestamp('today') - pd.offsets.DateOffset(years=1)).strftime('%Y%m%d'))
            prices.set_index('date', inplace=True)
            close_prices[ticker] = prices['close']
        except:
            pass
    returns = close_prices.pct_change()
    momentum = returns.iloc[-1]  # 取最近一天的收益率作为动量因子
    momentum_rank = momentum.rank(pct=True)  # 将动量因子归一化为排名
    return momentum_rank

# 选取强势证券
def select_stocks(momentum_rank, n):
    return momentum_rank[momentum_rank > (1 - n)]

# 回测投资组合
def backtest(selected_stocks, start_date, end_date):
    prices = pd.DataFrame()
    for ticker in selected_stocks.index.tolist():
        prices[ticker] = ak.stock_zh_a_daily(symbol=ticker, adjust="hfq", start_date=start_date, end_date=end_date)['close']
    returns = prices.pct_change().dropna()
    equal_weighted_returns = returns.mean(axis=1)
    return equal_weighted_returns.cumsum()
```

```python
# 运行策略
momentum_rank = momentum_factor(tickers)
momentum_rank
```

```python
selected_stocks = select_stocks(momentum_rank, 0.2)  # 选取排名前20%的强势证券
portfolio_returns = backtest(selected_stocks, start_date='2015-01-01', end_date='2020-12-31')

# 输出回测结果
print('投资组合年化收益率:', (portfolio_returns[-1]+1)**(252/len(portfolio_returns))-1)
print('投资组合夏普比率:', portfolio_returns.mean() / portfolio_returns.std() * np.sqrt(252))
```
